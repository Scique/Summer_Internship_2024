```python
import numpy

# Open dataset
data = open("./dataset/seattle-weather.csv", "r")

# First line is the feature names
# No need for weather since weather will be the target names
featureNames = data.readline()[:-1].split(",")[:-1]

# Labels for each dataset
targetNames = numpy.array(["drizzle", "rain", "snow", "fog"])

# Turn csv into array format
sortedData = []
targets = []
for line in data:
    # Remove /n from the end of the line
    line = line[:-1]
    # Get target names (the last section of code)
    lineList = line.split(",")
    # Convert to numeric values
    # Convert date
    lineList[0] = "".join([char for char in lineList[0] if char.isdigit()])
    # Add targets
    targets.append(lineList[-1])
    # Convert the rest
    # Skip the weather
    lineList = [float(value) for value in lineList[:-1]]
    sortedData.append(lineList)

# Turn targetNames into numpy array
targets = numpy.array(targets)
sortedData = numpy.array(sortedData)

print(targets)
print(sortedData)
print(featureNames)

# Array first axis = samples, second axis = features  

# Split the data into a 70-30 train to test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(sortedData, targets, test_size = 0.3, random_state = 1)

# Verify shape
print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)





```

Predict the data

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
# Finding accuracy by comparing actual response values(y_test)with predicted response value(y_pred)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
# Providing sample data and the model will make prediction out of that data
from sklearn import svm
svc = svm.SVC(kernel='sigmoid')
svc.fit(X_train, y_train)

y_pred = svc.predict(X_test)

print(f"Accuracy SVC: {metrics.accuracy_score(y_test, y_pred)}")

```
